# Phase 2: Building Machine Learning Models

In this phase, we will load our perfectly cleaned datasets from Phase 1 and build models to predict Customer Churn.

We will train identical models on two different versions of our data to see which performs better:
1. **MinMax Scaled Data:** Features are squished between 0 and 1.
2. **Standard Scaled Data:** Features are centered around a mean of 0 with a standard deviation of 1.

```python
# 1. Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

## 1. Load the Data
Let's load both of the CSV files we created in Phase 1.

```python
# Load MinMax Scaled Data
df_minmax = pd.read_csv('Telco-Customer-Churn-MinMax.csv')

# Load Standard Scaled Data
df_standard = pd.read_csv('Telco-Customer-Churn-Standard.csv')

print(f"MinMax Data Shape: {df_minmax.shape}")
print(f"Standard Data Shape: {df_standard.shape}")
```

## 2. Train-Test Split
Before training, we must split our data. 
* **X (Features):** Everything *except* the Churn column.
* **y (Target):** Only the Churn column.

We use `train_test_split` to randomly divide our data: 80% for training the model, and 20% reserved for testing its accuracy.

```python
# Define targets and features for MinMax Data
X_minmax = df_minmax.drop('Churn', axis=1)
y_minmax = df_minmax['Churn']

# Split: 80% Train, 20% Test
X_train_minmax, X_test_minmax, y_train_minmax, y_test_minmax = train_test_split(
    X_minmax, y_minmax, test_size=0.2, random_state=42
)

print("MinMax Split Done!")
```

```python
# Define targets and features for Standard Data
X_std = df_standard.drop('Churn', axis=1)
y_std = df_standard['Churn']

# Split: 80% Train, 20% Test
X_train_std, X_test_std, y_train_std, y_test_std = train_test_split(
    X_std, y_std, test_size=0.2, random_state=42
)

print("Standard Split Done!")
```

## 3. Train the Models (Logistic Regression)
Now we will train two separate Logistic Regression models: one predicting on the MinMax data, and one on the Standard scaled data.

```python
--- Model 1: MinMax Scaled Data ---
# 1. Initialize the Model
model_minmax = LogisticRegression()

# 2. Train the Model using the training data
model_minmax.fit(X_train_minmax, y_train_minmax)

# 3. Make predictions on the hidden test data
predictions_minmax = model_minmax.predict(X_test_minmax)

# 4. Check Accuracy
acc_minmax = accuracy_score(y_test_minmax, predictions_minmax)
print(f"Accuracy using MinMaxScaler: {acc_minmax * 100:.2f}%")
```

```python
--- Model 2: Standard Scaled Data ---
# 1. Initialize the Model
model_std = LogisticRegression()

# 2. Train the Model using the training data
model_std.fit(X_train_std, y_train_std)

# 3. Make predictions on the hidden test data
predictions_std = model_std.predict(X_test_std)

# 4. Check Accuracy
acc_std = accuracy_score(y_test_std, predictions_std)
print(f"Accuracy using StandardScaler: {acc_std * 100:.2f}%")
```
